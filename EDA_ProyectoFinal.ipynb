#Análisis de la plataforma streaming CrunchyRoll de mayo del 2022 en EEUU

CrunchyRoll es una plataforma de streaming especializada en Anime que ha ganado mucha popularidad debido a la facilidad de ver esta clase de contenido de forma segura, traducción oficial y aparte permite a los usuarios estar actualizados poco tiempo después de los estrenos en Japón.

La plataforma se fundó en 2006.

![Logo de CrunchyRoll](https://upload.wikimedia.org/wikipedia/commons/1/1f/%D8%B4%D8%B9%D8%A7%D8%B1_%D9%83%D8%B1%D8%A7%D9%86%D8%B4%D9%8A_%D8%B1%D9%88%D9%84.png)

Este dataset contienen todos los animes disponibles en la plataforma, con sus titulos y demás información relevante, como dice el título los datos constan del período de mayo de 2022, toda la data fue recolectada de la plataforma JustWatch que ordena la lista de contenido de distintas plataformas de streaming por su popularidad en los Estados Unidos.

Meta datos:
 * id: The title ID on JustWatch
 * title: The name of the title.
 * show_type: TV show or movie.
 * description: A brief description.
 * release_year: The release year.
 * age_certification: The age certification.
 * runtime: The length of the episode (SHOW) or movie.
 * genres: A list of genres.
 * production_countries: A list of countries that   produced  the title.
 * seasons: Number of seasons if it's a SHOW.
 * imdb_id: The title ID on IMDB.
 * imdb_score: Score on IMDB.
 * imdb_votes: Votes on IMDB.
 * tmdb_popularity: Popularity on TMDB.
 * tmdb_score: Score on TMDB.

```python
columna = df['imdb_score']

plt.hist(columna, bins=10)
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de imdb_score')

media = columna.mean()
mediana = columna.median()
moda = columna.mode()[0]

Q1 = columna.quantile(0.25)
Q3 = columna.quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
outliers = columna[(columna < limite_inferior) | (columna > limite_superior)]

plt.axvline(media, color='r', linestyle='--', label='Media')
plt.axvline(mediana, color='g', linestyle='-', label='Mediana')
plt.axvline(moda, color='b', linestyle=':', label='Moda')
plt.axvline(limite_inferior, color='y', linestyle='-.', label='Límite inferior outliers')
plt.axvline(limite_superior, color='y', linestyle='-.', label='Límite superior outliers')

plt.legend()
plt.show()

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)
print("Outliers:", outliers)
```

Los valores de la media, mediana y moda son muy cercanos mientras los outliers son valores bajos, es ligeramente asimétrica hacia la izquierda, para que no se vean afectado la sustitución es mejor usar la mediana.

```python
mediana = df['imdb_score'].median()

df['imdb_score'].fillna(mediana, inplace=True)
```
